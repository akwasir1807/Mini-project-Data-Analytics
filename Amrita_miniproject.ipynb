```python
from faker import Faker
```

```python
# Generate a fake dataset with 100 rows of random names, addresses, and dates using the Faker library
fake = Faker()
fake_data = pd.DataFrame({
    'Name' : [fake.name() for _ in range(100)],
    'Address' : [fake.address() for _ in range(100)],
    'Date' : [fake.date() for _ in range(100)],
})
```

```python
fake_data
```

```python
# Convert 'Date_of_Birth' to Age
from datetime import datetime

fake_date = pd.DataFrame({
    'Date_of_Birth' : [fake.date_of_birth(minimum_age =18, maximum_age =80 )  for _ in range(100)]
})
fake_date
```

```python
fake_date['Age'] = fake_date['Date_of_Birth'].apply(lambda x: datetime.now().year - x.year)
fake_date
```

```python
# Merge two DataFrames on a common column 'Product_ID'
df1 = pd.DataFrame({
    "Product_ID": [101, 102, 103, 104],
    "Product_Name": ["Laptop", "Phone", "Tablet", "Monitor"],
    "Price": [800, 500, 300, 250]
})

df2 = pd.DataFrame({
    "Product_ID": [102, 103, 104, 105],
    "Stock": [50, 30, 20, 10]
})

merged_df = pd.merge(df1, df2, on="Product_ID")
merged_df
```
